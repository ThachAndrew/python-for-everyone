## Reduction

A reduction algorithm takes a sequence of values and generates a single value from them. Finding the sum of all the numbers in a list, their average or the minimum value are examples of reduction algorithms. 

![The reduction algorithm](images/alg-reduce.svg)

A reduction algorithm start by initializing an output variable. During each step of the `for` loop, the output variable is updated with a new value. After the `for` loop is complete a reduction algorithm can optionally process or adjust the output variable. Here's an example of an algorithm that calculates the average of all the numbers in a list: 

```python
def average(numbers):
    output = 0 # Initialize the output variable 
    for number in numbers:
        output = output + number   # Update the output variable 
    output = output / len(numbers) # Adjust the final value
    return output
```

Let's simulate in our head what the `less_than_20` function does:


```python 
less_than_20([1, 23, 46, 92, 9, 12, 5])
```

Notice that the `if` is inside the `for`. Simulating the loop in a table gives us: 

| Step # | `number` | `number < 20` | `output` | 
| --- | --- | --- | --- | 
| start | - | - | `[]` | 
| 1 | `1` | `True` | `[1]` |  
| 2 | `23` | `False` | `[1]` |  
| 3 | `46` | `False` | `[1]` |  
| 4 | `92` | `False` | `[1]` |  
| 5 | `9` | `True` | `[1, 9]` |  
| 6 | `12` | `True` | `[1, 9, 12]` |  
| 7 | `5` | `True` | `[1, 9, 12, 5]` |  

Now enter the example and use the debugger to confirm our thinking:

## Mapping 

A mapping algorithm applies an operation to every element in a sequence, creating a new sequence. 

![The mapping algorithm](images/alg-map.svg)

Mapping algorithms start with an empty output and append a new value to it every time through the `for` loop. The value that's appended has the operation applied. Here's an example that doubles each value in an input list to produce an output list:

```python 
def double_list(numbers):
    output = [] # Start with an empty output
    for number in numbers:
        output.append(number * 2) # Apply the operation to every element.
    return output 
```

Let's simulate what the `double_list` function does: 

```python

```

Simulating the loop in a table gives us: 

| Step # | `number` | `output` | 
| --- | --- | --- | 
| start | - | `[]` | 
| 1 | `1`  | `[2]` |  
| 2 | `23`  | `[2, 46]` |  
| 3 | `46`  | `[2, 46, 92]` |  
| 4 | `92`  | `[2, 46, 92 184]` |  
| 5 | `9`  | `[2, 46, 92, 184, 18]` |  
| 6 | `12`  | `[2, 46, 92, 184, 18, 24]` |  
| 7 | `5`  | `[2, 46, 92, 184, 18, 24, 10]` |  


## Example: Searching for a Value 

Sometimes you want to find something in the list. Of course, this is conveniently done with the `in` operator in Python but let's see it in a `for` loop. A search algorithm is a *reduction* algorithm because it takes a sequence as input and returns a single value, `True` or `False`. Let's take a look at a simplistic search:

```python
def search(value, numbers):
    output = False 
    for number in numbers:
        if number == value:
            output = True
    return output
```

Let's simulate the algorithm with this input: 

```python 
search(23, [1, 23, 46, 92, 9, 12, 5])
```


| Step # | `number` | `output` | 
| --- | --- | --- | 
| start |  - | `False` | 
| 1 | `1` | `False` | 
| 2 | `23` | `True` | 
| 3 | `46` | `True` | 
| 4 | `92` | `True` | 
| 5 | `9` | `True` | 
| 6 | `12` | `True` | 
| 7 | `5` | `True` | 
| last | - | `True` | 

This algorithm is wasteful because it kept looking after it had already found `23`. An algorithm that searches a sequence should stop when it has found what it's looking for. The `break` statement stops a for loop from executing, leaving any remaining items unseen. Here's an algorithm that searches for a number in the list:

```python
def search(value, numbers):
    output = False 
    for number in numbers:
        if number == value:
            output = True
            break
    return output
```

Running `break` stops the loop early. Here's a simulation: 

| Step # | `number` | `output` | 
| --- | --- | --- | 
| start |  - | `False` | 
| 1 | `1` | `False` | 
| 2 | `23` | `True` | 
| last | - | `True` | 

<div class="alert alert-info"><strong>Note: Programming Style</strong>
<br><br>
<p>The break statement is controversial. Some people discourage its use because they believe that it makes code <strong>less readable</strong>, or harder to simulate in your head. As you gain experience you'll be able to judge for yourself how readable and clear code is. Every statement has a time and a place.
</div>

Let's rewrite `search` without the `break`:

```python 
def search(value, numbers):
    for number in numbers:
        if number == value:
            return True
    return False
```


Enter the example and confirm the simulation in the debugger:
