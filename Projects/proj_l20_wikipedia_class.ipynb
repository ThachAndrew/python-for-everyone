# Wikipedia Helper 

In this project you will use a Python class to create an Appilcation Programming Interface (API) to help access Wikipedia. You can get programmatic data in the [JSON](https://en.wikipedia.org/wiki/JSON) format. JSON is easy to translate into Python dictionaries.  


## The Wikipedia Class

Your Wikipedia class should have the following definition: 

- Name: `Wikipedia`
- Functions:
  - `__init__` - The class initializer
    - Arguments:
      - `self`: The class instance 
      - `page`: The title of a page to access. 
  - `exists` - Returns `True` if the page exists, `False` otherwise. 
    - Arguments: `None`
    - Returns: `True` or `False`
  - `extract` - Return the extract of the page. 
    - Arguments: `None`
    - Returns: The page extract as a string. 
    
## Getting the Page Information

Getting page information from Wikipedia is easy (if you know how to do it). The `wiki_fetch.py` program in Lesson11 shows you how. The program shows you the dictionary that's returned by Wikipedia. You should look at the output closely to learn the structure.
