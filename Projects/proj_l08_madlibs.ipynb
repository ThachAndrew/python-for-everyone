# Project: Mad Libs  

In this project you'll take user input to create a Mad Lib generator. Haven't heard of Mad Libs? It's a word game: 

http://www.madlibs.com/

Mad Libs are like templates. Here's a PDF of some typical Mad Libs: 

http://www.madlibs.com/content/uploads/2016/04/KIDS-excuse-hall-pass_1_.pdf

One person knows the template and asks for particular types of words (e.g. noun, adjective). The player that picks the words doesn't know the template. At the end the template is read with the randomly picked words and hilarity ensues. 

## A Madlib Generator 

You program will be called with one command line input. The input will be the Madlib template. The template will have three missing pieces, a noun a verb and an adjective. Here's an example of how your program will be called: 

```bash
$ python3.6 ./mad_lib.py 'I like to {verb} {adjective} {noun}.'
```

Once called your program must prompt the user for the three missing words. The words will always be noun, verb and adjective and you should prompt the user in that order: 

```bash
$ python3.6 ./mad_lib.py 'I like to {verb} {adjective} {noun}.'
Please enter a noun:
Please enter a verb:
Please enter an adjective:
```

After gathering three words your program prints the completed Madlib. For example if the user entered the words, "word", "ran" and "rock" the program execution would look like this: 

```bash
$ python3.6 ./mad_lib.py 'I like to {verb} {adjective} {noun}.'
Please enter a noun: word
Please enter a verb: ran
Please enter an adjective: blue
I like to ran blue word
```

## Algorithm 

An *algorithm* is a procedure for doing something in a program, a sequence of steps. This program is an algorithm with the steps: 

  1. Get the command line input. 
  2. Ask the user for a `noun` 
  3. Ask the user for a `verb` 
  4. Ask the user for an `adjective` 
  5. Print the completed Mad Lib 

Think about the code to implement these steps, and assemble the code in order. 

## Quotes on the UNIX Command Line 

You must enclose your command line arguments in single-quotes `'` or else the shell will be unhappy!

**This is correct:**

```bash 
$ python3.6 ./mad_lib.py 'I like to {verb} {adjective} {noun}.'
```

**This is incorrect:**

```bash
$ python3.6 ./mad_lib.py I like to {verb} {adjective} {noun}.
```

## Requirements 

  - Your program must have the required docstrings 
  - You must take exactly one command line argument 
  - You must prompt the user for three inputs (verb, adjective, noun) in order
  - You must print the completed madlib after taking input 
  - Your project must be in a file named `mad_lib.py`
